```python
problem_6 = ConcreteModel()

# define set A from 1 to 4
problem_6.A = RangeSet(1,4)

# define set C from 1 to 4
problem_6.C = RangeSet(1,4)

# define parameter e for each element in a and c
problem_6.e = Param(problem_6.A, problem_6.C, initialize={ (1,1): 6, (1,2): 2, (1,3): 8, (1,4): 5,
                                                            (2,1): 9, (2,2): 3, (2,3): 5, (2,4): 8,
                                                            (3,1): 4, (3,2): 8, (3,3): 3, (3,4): 4,
                                                            (4,1): 6, (4,2): 7, (4,3): 6, (4,4): 4})

# define variable x binary for each element in a and c
# var x{A,C} binary;

problem_6.x = Var(problem_6.A, problem_6.C, within=Binary)

# define objective function sum i in a, j in c e[i,j] * x[i,j]
problem_6.z = Objective(expr=sum(problem_6.e[i,j] * problem_6.x[i,j] for i in problem_6.A for j in problem_6.C), sense=maximize)

# define constraint sum i in a x[i,j] = 1 for each j in c
problem_6.r1 = ConstraintList()
for j in problem_6.C:
    problem_6.r1.add(sum(problem_6.x[i,j] for i in problem_6.A) == 1)

# define constraint sum j in c x[i,j] = 1 for each i in a
problem_6.r2 = ConstraintList()
for i in problem_6.A:
    problem_6.r2.add(sum(problem_6.x[i,j] for j in problem_6.C) == 1)

solver = SolverFactory('gurobi')
results = solver.solve(problem_6)

print('Solución optima:')
for x in problem_6.x:
    print(x, value(problem_6.x[x]))
print('Valor óptimo:', value(problem_6.z))
```

```python
problem_7 = ConcreteModel()

# define set I from 1 to 3
problem_7.I = RangeSet(1,3)

# define set J from 1 to 6
problem_7.J = RangeSet(1,6)

# define parameter c{I,J}
problem_7.c = Param(problem_7.I, problem_7.J, initialize={ (1,1): 13150, (1,2): 9650, (1,3): 19950, (1,4): 5000, (1,5): 9650, (1,6): 15950,
                                                            (2,1): 19950, (2,2): 6500, (2,3): 19950, (2,4): 9650, (2,5): 11150, (2,6): 15950,
                                                            (3,1): 17150, (3,2): 15950, (3,3): 19950, (3,4): 19950, (3,5): 15950, (3,6): 7000})

# offer dictionary
offer = {1: 600, 2: 600, 3: 250}

# define parameter o{I}
problem_7.o = Param(problem_7.I, initialize=offer)

# demand dictionary
demand = {1: 280, 2: 250, 3: 240, 4: 170, 5: 210, 6: 300}

# define parameter d{J}
problem_7.d = Param(problem_7.J, initialize=demand)

# question 1 is the problem viable with the offer and demand?
# take the sum of the offer and demand and compare
if sum(offer.values()) == sum(demand.values()):
    print('El problema es viable')
else:
    print('El problema no es viable')

# define variable x positive integer for each element in i and j
problem_7.x = Var(problem_7.I, problem_7.J, domain=NonNegativeIntegers)

# define objective function sum i in I, j in J c[i,j] * x[i,j]
problem_7.z = Objective(expr=sum(problem_7.c[i,j] * problem_7.x[i,j] for i in problem_7.I for j in problem_7.J), sense=minimize)

# define offer constraint sum j in J x[i,j] <= o[i] for each i in I
problem_7.r1 = ConstraintList()
for i in problem_7.I:
    problem_7.r1.add(sum(problem_7.x[i,j] for j in problem_7.J) <= problem_7.o[i])

# define demand constraint sum i in I x[i,j] >= d[j] for each j in J
problem_7.r2 = ConstraintList()
for j in problem_7.J:
    problem_7.r2.add(sum(problem_7.x[i,j] for i in problem_7.I) >= problem_7.d[j])

solver = SolverFactory('gurobi')
results = solver.solve(problem_7)

print('Solución optima:')
for x in problem_7.x:
    print(x, value(problem_7.x[x]))
print('Valor óptimo:', value(problem_7.z))
```

```python
problem_8 = ConcreteModel()

# define set I from 1 to 3
problem_8.I = RangeSet(1,3)

# define set J from 1 to 4
problem_8.J = RangeSet(1,4)

# investment cost dictionary
investment_cost = {1: 250000, 2: 180000, 3: 70000}

# define parameter C{I}
problem_8.C = Param(problem_8.I, initialize=investment_cost)

# sell price dictionary
sell_price = {1: 400000, 2: 270000, 3: 130000}

# define parameter PV{I}
problem_8.PV = Param(problem_8.I, initialize=sell_price)

# define variable x positive integer for each element in i and j
problem_8.x = Var(problem_8.I, problem_8.J, domain=NonNegativeIntegers)

# define objective function sum i in I, j in J PV[i] * x[i,j] - C[i] * x[i,j]
problem_8.z = Objective(expr=sum(problem_8.PV[i] * problem_8.x[i,j] - problem_8.C[i] * problem_8.x[i,j] for i in problem_8.I for j in problem_8.J), sense=maximize)

# global budget
budget = 3000000

# define constraint sum{i in I, j in J} C[i]*x[i,j] <= budget
problem_8.r1 = ConstraintList()
problem_8.r1.add(sum(problem_8.C[i] * problem_8.x[i,j] for i in problem_8.I for j in problem_8.J) <= budget)

# maximum number of items through customs
max_customs = 12

# define constraint {j in J} sum i in I x[i,j] <= max_customs
problem_8.r2 = ConstraintList()
for j in problem_8.J:
    problem_8.r2.add(sum(problem_8.x[i,j] for i in problem_8.I) <= max_customs)

solver = SolverFactory('gurobi')
results = solver.solve(problem_8)

print('Solución optima:')
for x in problem_8.x:
    print(x, value(problem_8.x[x]))
print('Valor óptimo:', value(problem_8.z))
```
